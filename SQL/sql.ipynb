## 1. Obter a lista das 10 cidades mais populosas em 2012

```python
query = '''SELECT ville_nom, ville_population_2012 
        FROM villes_france_free 
        ORDER BY ville_population_2012 
        DESC 
        LIMIT 10'''
pd.read_sql_query(query,conn)
```

## 2. Obter a lista das 50 cidades de menor superfície

```python
query = '''SELECT ville_nom, ville_surface 
        FROM villes_france_free 
        ORDER BY ville_surface 
        ASC 
        LIMIT 50'''
pd.read_sql_query(query,conn)
```

## 3. Obter a lista de departementos de "outre-mer" (aqueles que começam com 97)

```python
query = '''SELECT * 
        FROM departement 
        WHERE (departement_code 
        LIKE '97%')'''
pd.read_sql_query(query,conn)
```

## 4. Obter o nome das 10 cidades mais populosas em 2012, assim como o nome do departamento associado

```python
query = '''SELECT v.ville_nom, v.ville_population_2012, d.departement_nom
        FROM villes_france_free as v
        INNER JOIN departement as d ON v.ville_departement=d.departement_code
        ORDER BY v.ville_population_2012
        DESC
        LIMIT 10;
        '''
pd.read_sql_query(query,conn)
```

```python
query = '''SELECT ville_nom, departement_nom, ville_population_2012
        FROM villes_france_free 
        JOIN departement
        ON villes_france_free.ville_departement = departement.departement_id 
        ORDER BY ville_population_2012
        DESC
        LIMIT 10'''
pd.read_sql_query(query,conn)
```

## 5. Obter a lista do nome de cada departamento, associado a seu código e ao número de cidades dentro dele, ordenando os departamentos em seguida por número de cidades

OBS: https://fr.wikipedia.org/wiki/Nombre_de_communes_par_d%C3%A9partement_en_France_au_1er_janvier_2014

```python
query = '''SELECT d.departement_code, d.departement_nom, COUNT(v.ville_id) as qtde_cidades
           FROM departement as d
           LEFT JOIN villes_france_free as v
           ON d.departement_code = v.ville_departement
           GROUP BY d.departement_code
           ORDER BY qtde_cidades
           DESC;
        '''
pd.read_sql_query(query,conn)
```

```python
query = '''SELECT departement_nom, departement_code, COUNT(departement_code) AS nombre_villes
        FROM departement
        JOIN villes_france_free
        ON departement.departement_code = villes_france_free.ville_departement
        GROUP BY departement_code
        ORDER BY nombre_villes DESC'''
pd.read_sql_query(query,conn)
```

## 6. Obter a lista de maiores departamentos, em termos de superfície

OBS: https://fr.wikipedia.org/wiki/Liste_des_d%C3%A9partements_fran%C3%A7ais_class%C3%A9s_par_population_et_superficie

```python
query = '''SELECT d.departement_nom, SUM(v.ville_surface) as departement_surface
           FROM departement as d
           LEFT JOIN villes_france_free as v
           ON d.departement_code = v.ville_departement
           GROUP BY d.departement_code
           ORDER BY departement_surface
           DESC;
        '''
pd.read_sql_query(query,conn)
```

```python
query = '''SELECT departement_nom, SUM(ville_surface) as departement_surface
        FROM departement
        JOIN villes_france_free
        ON departement.departement_code = villes_france_free.ville_departement
        GROUP BY departement_code
        ORDER BY departement_surface DESC'''
pd.read_sql_query(query,conn)
```

## 7. Contar o número de cidades cujo nome começa por "Saint"

```python
query = '''SELECT COUNT(ville_nom) as count_saint
        FROM villes_france_free
        WHERE ville_nom LIKE "Saint%"'''
pd.read_sql_query(query,conn)
```

## 8. Obter a lista de cidades que possuem um nome existente mais de uma vez, e ordená-las para obter em primeiro lugar aquelas cujo nome é mais utilizado por outras cidades

```python
query = '''SELECT ville_nom, count(ville_nom) as count_ville_nom
        FROM villes_france_free
        GROUP BY ville_nom
        ORDER BY count_ville_nom DESC
        LIMIT 10'''
pd.read_sql_query(query,conn)
```

## 9. Obter em uma mesma requisição SQL a lista das cidades cuja superfície é superior à superfície média

```python
query = '''SELECT ville_nom, ville_surface
        FROM villes_france_free
        WHERE ville_surface > (
            SELECT AVG(ville_surface) FROM villes_france_free
        )'''
pd.read_sql_query(query,conn)
```

## 10. Obter a lista de departamentos que possuem mais de 2 milhões de habitantes

```python
query = '''SELECT d.departement_id, d.departement_nom, SUM(v.ville_population_2012) as sum_dept_pop_2012
           FROM departement as d
           LEFT JOIN villes_france_free as v
           ON d.departement_id = v.ville_departement
           GROUP BY d.departement_code
           HAVING sum_dept_pop_2012 > 2000000;
        '''
pd.read_sql_query(query,conn)
```

```python
query = '''SELECT departement_nom, departement_code, SUM(ville_population_2012) AS departement_population
        FROM departement
        JOIN villes_france_free
        ON departement.departement_code = villes_france_free.ville_departement
        GROUP BY departement_code
        HAVING departement_population > 2000000'''
pd.read_sql_query(query,conn)
```

## 11. Substituir os "-" por um espaço em branco, para todas as cidades que começam por "SAINT-"

```python
query = '''UPDATE villes_france_free
        SET ville_nom = REPLACE(ville_nom, '-', ' ')
        WHERE ville_nom LIKE "SAINT%"'''
cursor.execute(query)
```

```python
query = '''SELECT *
        FROM villes_france_free
        WHERE ville_nom LIKE "SAINT%"'''
pd.read_sql_query(query,conn)
```

M√©tricas de regress√£o

```python
from sklearn import metrics

def calcMetricas(y, y_pred, verbose=False):
    metricas = {}
    metricas['mae'] = metrics.mean_absolute_error(y, y_pred)
    metricas['mse'] = metrics.mean_squared_error(y, y_pred)
    metricas['rmse'] = np.sqrt(metrics.mean_squared_error(y, y_pred))
    metricas['r2'] = metrics.r2_score(y, y_pred)
    
    if verbose:
        print('Mean Absolute Error:', metricas['mae'])
        print('Mean Squared Error:', metricas['mse'])
        print('Root Mean Squared Error:', metricas['rmse'])
        print('R2 score:', metricas['r2'])
    
    return metricas
```

## DataPrep

Vamos utilizar somente as variaveis num√©ricas, a principio

```python
numeric_fields = ['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces', 'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold']
house_train = house_train[numeric_fields + ['SalePrice']].copy()
house_test = house_test[numeric_fields].copy()
```

Vamos fazer a substitui√ß√£o de NaN's por zeros

```python
house_train.fillna(0, inplace=True)
house_test.fillna(0, inplace=True)
```

Vamos dividir o dataset de treino em treino e valida√ß√£o

```python
X_train = house_train.drop(columns=['SalePrice'])
y_train = house_train['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_valid.shape, y_train.shape, y_valid.shape
```

## Cross Validation

Vamos utilizar K-Fold CV

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression

kf = KFold(n_splits=5, shuffle=True, random_state=42)
metricas_treino = []
metricas_teste = []
count = 1
for train_index, test_index in kf.split(X_train):
    print("========== RODADA {} ==========".format(count))
    print("TRAIN:", len(train_index), "TEST:", len(test_index))
    
    X_cv_train, X_cv_test = X_train.iloc[train_index], X_train.iloc[test_index]
    y_cv_train, y_cv_test = y_train.iloc[train_index], y_train.iloc[test_index]
    
    print("# Fazendo fit do modelo")
    model = LinearRegression(normalize=True)
    model.fit(X_cv_train, y_cv_train)
    
    print("# Fazendo predi√ß√µes")
    y_cv_pred_train = model.predict(X_cv_train)
    y_cv_pred_test = model.predict(X_cv_test)
    
    print("# Calculando as m√©tricas")
    print("## Treino")
    metricas_cv_treino = calcMetricas(y_cv_train, y_cv_pred_train, verbose=True)
    print("## Teste")
    metricas_cv_teste = calcMetricas(y_cv_test, y_cv_pred_test, verbose=True)
    print()
    
    metricas_treino.append(metricas_cv_treino)
    metricas_teste.append(metricas_cv_teste)
    
    count += 1

print("================ RESULTADO ================")
print("## Treino")
print("M√©dia do MAE", np.mean(list(map(lambda x: x['mae'], metricas_treino))))
print("M√©dia do MSE", np.mean(list(map(lambda x: x['mse'], metricas_treino))))
print("M√©dia do RMSE", np.mean(list(map(lambda x: x['rmse'], metricas_treino))))
print("M√©dia do R2", np.mean(list(map(lambda x: x['r2'], metricas_treino))))
print()
print("## Teste")
print("M√©dia do MAE", np.mean(list(map(lambda x: x['mae'], metricas_teste))))
print("M√©dia do MSE", np.mean(list(map(lambda x: x['mse'], metricas_teste))))
print("M√©dia do RMSE", np.mean(list(map(lambda x: x['rmse'], metricas_teste))))
print("M√©dia do R2", np.mean(list(map(lambda x: x['r2'], metricas_teste))))
```

## Treinando o modelo

Vamos utilizar um modelo simples de Regress√£o Linear

```python
model = LinearRegression(normalize=True)
model.fit(X_train, y_train)
```

## Testando o resultado

Aplicamos train e valida√ß√£o no modelo treinado e vemos as m√©tricas

```python
y_pred_train = model.predict(X_train)
y_pred_valid = model.predict(X_valid)

print("# Calculando as m√©tricas")
print("## Treino")
metricas_treino = calcMetricas(y_train, y_pred_train, verbose=True)
print("## Valida√ß√£o")
metricas_valid = calcMetricas(y_valid, y_pred_valid, verbose=True)
```

```python
X_test = house_test.copy()
y_pred_test = model.predict(X_test)
```
